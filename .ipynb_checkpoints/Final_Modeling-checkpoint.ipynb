```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
import seaborn as sns
sns.set_style('darkgrid')
```

```python
WEST = pd.read_excel('WEST.xlsx').set_index('index').fillna(method='ffill')
MIDATL = pd.read_excel('MIDATL.xlsx').set_index('index').fillna(method='ffill')
SOUTH = pd.read_excel('SOUTH.xlsx').set_index('index').fillna(method='ffill')
dfs = [WEST,MIDATL,SOUTH]
```

```python
for i in range(len(dfs)):
    dfs[i].index = pd.to_datetime(dfs[i].index).round('h')
    dates = dfs[i].index
    dfs[i]['MonthNumber'] = dfs[i].index.month
    dfs[i]['WeekNumber'] = dfs[i].index.week
    dfs[i]['DayNumber'] = dfs[i].index.dayofweek
    dfs[i]['HourNumber'] = dfs[i].index.hour
    columns = dfs[i].columns
    x = dfs[i].values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    dfs[i] = pd.DataFrame(x_scaled, columns = columns)
    dfs[i].index = dates

WEST = dfs[0]
MIDATL = dfs[1]
SOUTH = dfs[2]
```

**BASELINE MODEL**


```python
plt.figure(figsize = (35,25))
dfs = [WEST,MIDATL,SOUTH]
regions = ['WEST','MIDATL','SOUTH']
REN = ['_solar','_wind','_demand']

names = ['West', 'Mid-Atlantic', 'South']
ranges = [range(1,4),range(4,7),range(7,10)]

lm = LinearRegression()

for r in range(len(REN)):
    
    for i,g in zip(range(len(dfs)) , ranges[r] ):
       
        ax = plt.subplot(3,3,g)
        
        X = dfs[i].drop(dfs[i].columns[-7:-4], axis=1)
        y = dfs[i][str(regions[i])+REN[r]]

        split = 0.99

        X_train, X_test, y_train, y_test = X[:int(split*len(X))], X[int(split*len(X)):], y[:int(split*len(X))], y[int(split*len(X)):]        
        
        
        lm.fit(X_train,y_train)
        
        linear_predict = lm.predict(X_test)
                
        plt.plot(X_test.index, y_test, label = 'Actual')
        plt.plot(X_test.index, linear_predict, label = 'Predicted')
        plt.legend(fontsize = 14)
        plt.ylabel('Normalized Solar Output', fontsize = 14);
        plt.xlabel('Date', fontsize = 14);

        if REN[r] == '_solar':
            plt.title('%s Solar Output Prediction' %(names[i]), fontsize=14)
            
         
        if REN[r] == '_wind':
            plt.title('%s Wind Output Prediction' %(names[i]), fontsize=14)
            
      
        if REN[r] == '_demand':
            plt.title('%s Demand Prediction' %(names[i]), fontsize=14)
```

**DECISION TREE** 

```python
plt.figure(figsize = (35,25))
dfs = [WEST,MIDATL,SOUTH]
regions = ['WEST','MIDATL','SOUTH']
REN = ['_solar','_wind','_demand']

names = ['West', 'Mid-Atlantic', 'South']
ranges = [range(1,4),range(4,7),range(7,10)]

tr = DecisionTreeRegressor()

for r in range(len(REN)):
    
    for i,g in zip(range(len(dfs)) , ranges[r] ):
       
        ax = plt.subplot(3,3,g)
        
        X = dfs[i].drop(dfs[i].columns[-7:-4], axis=1)
        y = dfs[i][str(regions[i])+REN[r]]

        split = 0.99

        X_train, X_test, y_train, y_test = X[:int(split*len(X))], X[int(split*len(X)):], y[:int(split*len(X))], y[int(split*len(X)):]        
        
        tr.fit(X_train,y_train)
        
        tree_predict = tr.predict(X_test)
                
        plt.plot(X_test.index, y_test, label = 'Actual')
        plt.plot(X_test.index, tree_predict, label = 'Predicted')
        plt.legend(fontsize = 14)
        plt.ylabel('Normalized Solar Output', fontsize = 14);
        plt.xlabel('Date', fontsize = 14);

        if REN[r] == '_solar':
            plt.title('%s Solar Output Prediction' %(names[i]), fontsize=14)
            
            print('%s Solar r-squared = ' %(regions[i]), r2_score(y_test, tree_predict))
            print('%s Solar RMSE =' %(regions[i]), mean_squared_error(y_test, tree_predict))
        if REN[r] == '_wind':
            plt.title('%s Wind Output Prediction' %(names[i]), fontsize=14)
            
            print('%s Wind r-squared = ' %(regions[i]), r2_score(y_test, tree_predict))
            print('%s Wind RMSE =' %(regions[i]), mean_squared_error(y_test, tree_predict))
        if REN[r] == '_demand':
            plt.title('%s Demand Output Prediction' %(names[i]), fontsize=14)
            
            print('%s Demand r-squared = ' %(regions[i]), r2_score(y_test, tree_predict))
            print('%s Demand RMSE =' %(regions[i]), mean_squared_error(y_test, tree_predict))
```

**RANDOM FOREST**

**EVALUATION**

```python
R_dt= np.zeros((3,3))
R_lr = np.zeros((3,3))
for r in range(len(REN)):
    
    for i,g in zip(range(len(dfs)) , ranges[r] ):
       
      
        
        X = dfs[i].drop([str(regions[i])+'_solar',str(regions[i])+'_demand',str(regions[i])+'_wind'], axis=1)
        Y = dfs[i][str(regions[i])+REN[r]]
        
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2, shuffle = False)
        
        dt = DecisionTreeRegressor(criterion = 'mse')
        dt.fit(X_train,y_train)
        y_1 = dt.predict(X_test)
        R_dt[r,i] = mean_squared_error(y_test, y_1, squared=False)
        
       
        lr = LinearRegression()
        lr.fit(X_train,y_train)
        y_2 = lr.predict(X_test)
        R_lr[r,i] = mean_squared_error(y_test, y_2, squared=False)
            
        
R_dt = pd.DataFrame(R_dt,index = regions , columns = REN)   
R_lr = pd.DataFrame(R_lr,index = regions , columns = REN)   
print('-------------------------------')
print('Decision Tree RMSE')
print('-------------------------------')
print(R_dt)


print('-------------------------------')
print('LR RMSE')
print('-------------------------------')
print(R_lr)

R_dt = R_dt.to_numpy()
R_lr = R_lr.to_numpy()
print('-----')
for i in range(3):
  a = np.sum(R_dt[:,i])
  b = np.sum(R_lr[:,i])

  if b > a:
    print('DT' + REN[i] +  ' RMSE % decrease vs LR')
    print(-100 * ((a-b)/b)) 
    print('-----')
  else:
    print('lr')
```
